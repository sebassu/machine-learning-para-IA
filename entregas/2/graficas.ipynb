```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n polin√≥mica de grado 2
def polinomio(x):
    return (x-1)**2

# Generar valores de x
x = np.linspace(-5, 5, 400)

# Calcular los valores correspondientes de y
y = polinomio(x)

# Crear la gr√°fica
plt.figure(figsize=(8, 8))
plt.plot(x, y, label='(w-1)^2')
plt.xlabel('x')
plt.ylabel('MSE(w)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xticks(np.arange(-5, 6, 1))
plt.legend()
plt.show()
```

```python
# 5.b

import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n polin√≥mica de grado 2
def polinomio(x):
    return 1/(1+x)

# Generar valores de x
x = np.linspace(-1, 5, 400)

# Calcular los valores correspondientes de y
y = polinomio(x)

# Crear la gr√°fica
plt.figure(figsize=(8, 8))
plt.plot(x, y, label='1/(1+alfa)')
plt.xlabel('alfa')
plt.ylabel('w')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xticks(np.arange(-2, 6, 1))
plt.legend()
plt.show()
```

```python
# 5.c

import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n polin√≥mica de grado 2
def polinomio(x):
    return (1-(7/(1+x)))**2

# Generar valores de x
x = np.linspace(0, 60, 400)

# Calcular los valores correspondientes de y
y = polinomio(x)

# Crear la gr√°fica
plt.figure(figsize=(8, 8))
plt.plot(x, y, label='(1-7/(1+alfa))^2')
plt.xlabel('alfa')
plt.ylabel('MSE')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
#plt.xticks(np.arange(-2, 101, 1))
plt.legend()
plt.show()
```

```python
# 5.e

import numpy as np
import matplotlib.pyplot as plt

# Definir la funci√≥n polin√≥mica de grado 2
def mse(x,a):
    return (x-1)**2+a*(x**2)

# Generar valores de x
x = np.linspace(-3, 3, 400)

alphas = np.arange(0.1, 10, 1)

for a in alphas:
    y = mse(x, a)
    plt.plot(x, y, label=f'a = {a}')

# Crear la gr√°fica
# plt.figure(figsize=(8, 8))
# plt.plot(x, y, label='(1-7/(1+alfa))^2')
plt.xlabel('alfa')
plt.ylabel('Ridge')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
#plt.xticks(np.arange(-2, 101, 1))
plt.legend()
plt.show()
```
